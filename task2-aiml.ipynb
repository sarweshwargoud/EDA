```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 5)


df = pd.read_csv("data/titanic.csv")
print("Shape:", df.shape)
display(df.head())
print(df.info())
display(df.describe(include="all"))
print("\nMissing values:\n", df.isnull().sum())


print("\nNumeric summary:")
display(df.describe())
print("\nCategorical summary:")
display(df.describe(include='object'))


df.hist(figsize=(12,8), bins=20)
plt.suptitle("Histograms of Numerical Features")
plt.show()


num_cols = ['Age','Fare']
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col], color='orange')
    plt.title(f"Boxplot of {col}")
    plt.show()


plt.figure(figsize=(10,6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()


sns.pairplot(df[['Survived','Pclass','Age','Fare']].dropna(), hue='Survived', vars=['Pclass','Age','Fare'])
plt.show()


plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=df, palette='Set2')
plt.title('Survival Count')
plt.show()

plt.figure(figsize=(8,5))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Pclass vs Survival')
plt.show()

plt.figure(figsize=(8,5))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Sex vs Survival')
plt.show()

plt.figure(figsize=(8,5))
sns.kdeplot(data=df, x='Age', hue='Survived', fill=True, common_norm=False)
plt.title('Age distribution by Survival')
plt.show()


df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
display(df[['FamilySize']].describe())

df['Title'] = df['Name'].str.extract(r',\s*([^\.]+)\.', expand=False).str.strip()
top_titles = df['Title'].value_counts().nlargest(10)
print("Top titles:\n", top_titles)

plt.figure(figsize=(8,4))
sns.boxplot(x='Title', y='Age', data=df[df['Title'].isin(top_titles.index)])
plt.xticks(rotation=45)
plt.title('Age by Title (top titles)')
plt.show()


print("\n--- Key Observations ---")
print("- Females had higher survival rates than males.")
print("- Passengers in Pclass 1 had higher survival.")
print("- Higher Fare roughly associates with higher survival probability.")
print("- Many ages are missing; consider imputation before modeling.")
print("- There are Fare outliers; use IQR or log-transform when modeling.")
print("- FamilySize and Title may be useful features for modeling.")


plt.figure(figsize=(8,5))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Pclass vs Survival')
plt.savefig('screenshots/pclass_survival.png', bbox_inches='tight')
plt.close()

plt.figure(figsize=(8,5))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Sex vs Survival')
plt.savefig('screenshots/sex_survival.png', bbox_inches='tight')
plt.close()

print("Saved example plot images to screenshots/")
```
